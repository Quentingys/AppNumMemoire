```python
"def gains"

%pylab inline 
import math 
import numpy as np
import matplotlib.pyplot as plt

"AllC = coopère tj"
"AllD = dénonce tj"
"Titfortat = puni si on le dénonce"
"DC = 1/2"

def pallc(strat, den, stratDC1):
    if strat == "AllC":
        G = 3
    elif strat == "AllD" :
        G = 0
    elif den == False and strat == "Titfortat" : 
        G = 3
    elif strat == "DC" and stratDC1 == "C":
        G = 3
    elif strat == "DC" and stratDC1 == "D":
        G=0
    return G 


def palld(strat, den, stratDC1):
    if strat == "AllC":
        G = 4 
    elif strat == "AllD":
        G = 1
    elif strat == "Titfortat" and den == False :
        G = 4 
    elif strat == "Titfortat" and den == True : 
        G = 1
    elif strat == "DC" and stratDC1 == "C" :
        G = 4
    elif strat == "DC" and stratDC1 == "D" : 
        G = 1
    return G

def ptft(strat, den, stratDC1):
    if strat == "AllC" :
        G = 3
    elif strat == "AllD" and den == False : 
        G = 0 
    elif strat == "AllD" and den == True : 
        G = 1
    elif strat == "Titfortat" :
        G = 3
    elif strat == "DC" and den == False : 
        G = 0
    elif strat == "DC" and den == True and stratDC1 == "C":
        G = 4 
    elif strat == "DC" and den == True and stratDC1 == "D":
        G = 1
    return G 

def pDC(strat, den, stratDC1, stratDC2) : 
    if strat == "AllC" and stratDC1 == "C":
        G = 3
    elif strat == "AllC" and stratDC1 == "D":
        G = 4
    elif strat == "AllD" and stratDC1 == "C":
        G = 0
    elif strat == "AllD" and stratDC1 == "D" : 
        G = 1
    elif strat == "Titfortat" and stratDC1 == "D" and den == False :
        G = 4 
    elif strat == "Titfortat" and stratDC1 == "D" and den == True : 
        G = 1 
    elif strat == "Titfortat" and stratDC1 == "C" and den == False :
        G = C
    elif strat == "Titfortat" and stratDC1 == "C" and den == True : 
        G = 0
    elif strat == "DC" and stratDC1 == "C" and stratDC2 == "C" : 
        G = 3
    elif strat == "DC" and stratDC1 == "C" and stratDC2 == "D" : 
        G = 0
    elif strat == "DC" and stratDC1 == "D" and stratDC2 == "C" : 
        G = 4
    elif strat == "DC" and stratDC1 == "D" and stratDC2 == "D" : 
        G = 1
    return G

print("All is ok")

   
    
```

```python
"Match"
def rencontre(strat1,strat2): 
    Nbcoups = 10
    "score AllC = Sallc"
    Sallc = 0
    "score AllD = Salld"
    Salld = 0
    "score Titfortat = Stft "
    Stft = 0
    "score DC = Sdc"
    Sdc = 0
    
    den = False
    i = 1
    stratDC1 = "D"
    stratDC2 = "D"
    
    while i <= Nbcoups : 
        
        if (strat1 == "AllC" and strat2 == "AllD") or (strat1 == "AllD" and strat2 == "AllC") : 
            Sallc = Sallc + pallc("AllD", den, stratDC1)
            Salld = Salld + palld("AllC", den, stratDC1)            
        elif (strat1 == "AllC" and strat2 == "Titfortat") or (strat1 == "Titfortat" and strat2 == "AllC"):
            Sallc = Sallc + pallc("Titfortat", den, stratDC1)
            Stft = Stft + ptft("AllC", den, stratDC1)
        elif (strat1 == "AllC" and strat2 == "DC") or (strat1 == "DC" and strat2 == "AllC") : 
            if i % 2 == 1 :
                stratDC1 = "D"
            elif i % 2 == 0 : 
                stratDC1 = "C"
            stratDC2 = stratDC1
            Sallc = Sallc + pallc("DC", den, stratDC1)
            Sdc = Sdc + pDC("AllC", den, stratDC1, stratDC2)
        elif (strat1 == "AllC" and strat2 == "AllC") :
            Sallc = Sallc + 2*pallc("AllC", den, stratDC1)
        
        
        elif (strat1 == "AllD" and strat2 == "Titfortat") or (strat1 == "Titfortat" and strat2 == "AllD"):
            if i > 1 : 
                den = True
            Salld = Salld + palld("Titfortat", den, stratDC1)
            Stft = Stft + ptft("AllD", den, stratDC1)
        elif (strat1 == "AllD" and strat2 == "DC") or (strat1 == "DC" and strat2 == "AllD"):
            if i % 2 == 1 :
                stratDC1 = "D"
            elif i % 2 == 0 : 
                stratDC1 = "C"
            stratDC2 = stratDC1
            Salld = Salld + palld("DC", den, stratDC1)
            Sdc = Sdc + pDC("AllD", den, stratDC1, stratDC2)
        elif (strat1 == "AllD" and strat2 == "AllD") :
            Salld = Salld + 2*palld("AllD", den, stratDC1)
            
    
        elif (strat1 == "DC" and strat2 == "Titfortat") or (strat1 == "Titfortat" and strat2 == "DC"):
            if i > 1 : 
                den = True
            if i % 2 == 1 :
                stratDC1 = "D"
            elif i % 2 == 0 : 
                stratDC1 = "C"
            stratDC2 = stratDC1
            Stft = Stft + ptft("DC", den, stratDC1)
            Sdc = Sdc + pDC("Titfortat", den, stratDC1, stratDC2)
        elif (strat1 == "Titfortat" and strat2 == "Titfortat"):
            Stft = Stft + 2*ptft("Titfortat", den, stratDC1)
        
        elif (strat1 == "DC" and strat2 == "DC"): 
            if i % 2 == 1 :
                stratDC1 = "D"
                stratDC2 = "D"
            elif i % 2 == 0 : 
                stratDC1 = "C"  
                stratDC2 = "C"
            Sdc = Sdc + 2*pDC("DC", den, stratDC1, stratDC2)
        i=i+1
            
    LS = [Sallc, Salld, Stft, Sdc]
    return LS

rencontre("AllC", "DC")
    
```
